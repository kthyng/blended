# Tidal influence at Galveston HF Radar locations

There is 6km spatial, hourly HF radar data available. 

```python
import netCDF4 as netCDF
import xarray as xr
import pandas as pd
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
loc = 'http://hfrnet.ucsd.edu/thredds/dodsC/HFR/USEGC/6km/hourly/RTV/HFRADAR,_US_East_and_Gulf_Coast,_6km_Resolution,_Hourly_RTV_best.ncd'
ds = xr.open_dataset(loc)
```

```python
ds
```

## Plot map


```python
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.Mercator(central_longitude=-85.0))
gl = ax.gridlines(linewidth=0.2, color='gray', alpha=0.5, linestyle='-', draw_labels=True)
ax.set_extent([-96, -93, 28.0, 29.9], ccrs.PlateCarree())
ax.coastlines(resolution='10m')  # coastline resolution options are '110m', '50m', '10m'
ax.contourf(ds['lon'], ds['lat'], ds['u'][-10,:,:], transform=ccrs.PlateCarree())
# ax.plot(ds['site_lon'], ds['site_lat'], 'o', transform=ccrs.PlateCarree())
```
